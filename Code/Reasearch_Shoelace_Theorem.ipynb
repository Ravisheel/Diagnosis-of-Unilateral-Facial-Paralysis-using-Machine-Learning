#Shoelace Theorem

```python
import matplotlib.pyplot as plt

# Define the coordinates of the four corners of the trapezoid
x = [30, 40, 60, 70]  # X-coordinates
y = [30, 50, 50, 30]  # Y-coordinates

# Point labels
labels = ['a', 'b', 'c', 'd']

# Plot the trapezoid
plt.plot(x + [x[0]], y + [y[0]], marker='o')

# Draw dotted lines from the origin to each point and label them
for i in range(len(x)):
    plt.plot([0, x[i]], [0, y[i]], 'r--')
    plt.text(x[i], y[i], labels[i], ha='right',fontsize=16)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')


plt.ylim(0, 80)
plt.xlim(0, 80)

# Show the plot
plt.grid(True)
plt.show()
```
