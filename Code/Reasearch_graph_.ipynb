```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = "Ratio of Area of Dental \nShow to Eye’s Area "

b = "Ratio of distance of \nCommissure to the \nFrankfurt Line "

c ="Nasolabial Fold’s depth "

d ="Smile Index "

e = "Eye Area "

f = "Mouth Area (Dental Show) "
```

```python
classifiers = [a,b,c,d,e,f,
    "combination",
    "logistic",
]

train_set_sizes = ["25%", "50%", "75%"]

# Accuracy data for each classifier and training set size
accuracy_data = {
    a: [69.23, 68, 70],
    b: [69.23, 76.52, 71.79],
    c: [71.79, 28.69, 76.62],
    d: [76.92, 85.21, 80.51],
    e: [71.79, 86.95, 81.81],
    f: [87.17, 82.6, 77.92],
    "combination": [70, 89.56, 90],
    "logistic": [92.3, 88.69, 86.03],
}

# Convert accuracy data to a NumPy array
accuracy_array = np.array([[accuracy_data[classifier][i] for classifier in classifiers] for i in range(3)])
```

```python
# ... (previous code)
bar_width = 0.3

# Generate the x-axis positions for each group
x = np.arange(len(classifiers))
# Create sub-bars for each training set size
for i, train_set_size in enumerate(train_set_sizes):
    plt.bar(
        x + i * bar_width,
        accuracy_array[i],
        bar_width,
        label=f"{train_set_size}",
    )

# Customize the plot
plt.xlabel("Classifiers")
plt.ylabel("Accuracy (%)")
plt.title("Accuracy of Different Classifiers for different test sizes")

# Rotate the x-axis tick labels horizontally
plt.xticks(x + bar_width, classifiers, rotation=90)  # Set rotation to 0 degrees for horizontal labels

plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.ylim(0,100)

# Show the plot
plt.tight_layout()
plt.show()
```
